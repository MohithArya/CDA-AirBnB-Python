### EDA

#### Data Analysis

#### Univariate Analysis

```python
# Price Distribution
plt.figure(figsize = (10,6))
plt.title("Price Distribution")
sns.histplot(df,x='price',bins=100,edgecolor='black')
plt.ylabel('FREQUENCY')
plt.xlabel('PRICE')
import warnings
warnings.filterwarnings('ignore')
```

```python
# availability_365
plt.figure(figsize = (8,5))
plt.title("Room Availability")
sns.histplot(df,x='availability_365',edgecolor='black',color='orange')
##plt.ylabel('FREQUENCY')
plt.xlabel('PRICE')
import warnings
warnings.filterwarnings('ignore')
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

#### Feature Engineering

```python
# Price Per Bed
df['Price_per_bed'] = df['price']/df['beds']
df.columns
```

```python
df[['Price_per_bed']]
```

```python
# Average Price per bed
df.groupby(by='neighbourhood_group')['Price_per_bed'].mean()
```

#### Bi-Variate Aanlysis

```python
# Price Dependency On neighbourhood
plt.title('Price Relativity with neighbourhood')
sns.barplot(df,x='neighbourhood_group',y='price',hue='room_type')
```

```python
# No. Of Review and price relative
plt.figure(figsize = (14,5))
plt.title('Locality And Review Relativity')
sns.scatterplot(df,x='number_of_reviews',y='price',hue='neighbourhood_group')
```

```python
# Pairplot
sns.pairplot(df, vars=['price','minimum_nights','number_of_reviews','availability_365',],hue='room_type')
```

```python
# Geographical Distribution
plt.figure(figsize=(12,6))
sns.scatterplot(df,x='longitude',y='latitude',hue='room_type')
plt.title('Geographical Distribution of AirBnb Listing')
plt.show()
```

```python
# HeatMap - corr of one variable with another for numerical columns
corr =df[['longitude','latitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap='coolwarm', linewidths=0.7)
plt.title('Correlation Between Numeric Values')
plt.show()
```
